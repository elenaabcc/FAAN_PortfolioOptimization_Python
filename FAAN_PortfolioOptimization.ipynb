```python
# Compute simple stock return which is = [(new price/old price) -1]- For totady is imposbile
returns=df.pct_change()
returns
```

```python
# Compute annualized covariance matrix
cov_matrix_annual = returns.cov() * 252
cov_matrix_annual
```

```python
# Choosing weights. How many stocks (in percentage) for each one.
weights = np.array([0.5, 0.2, 0.2, 0.1])
# 20% of my portfolio consists of Apple stock
```

```python
#let's find portfolio variance
portfolio_variance= np.dot(weights.T, np.dot(cov_matrix_annual, weights) )
portfolio_variance

```

```python
# Standard deviation known as portfolio volatility
portfolio_volatility = np.sqrt(portfolio_variance)
portfolio_volatility
```

```python
#find the annual portfolio return
portfolioAnnualReturn = np.sum(returns.mean() * weights) * 252
portfolioAnnualReturn
```

```python
# Try to calculate the expected annual return, volatility (risk) and variance
percent_var = str(round(portfolio_variance, 2 )*100) + "%"
percent_vol = str(round(portfolio_volatility, 2 )*100) + "%"
percent_annualreturn = str(round(portfolioAnnualReturn, 2 )*100) + "%"

print("Portfolio Variance :", percent_var)
print("portgolio Volatility", percent_vol)
print("Portfolio Annual Return",percent_annualreturn)
```
